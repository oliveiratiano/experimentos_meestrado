```python
import string
import unidecode
import re
import nltk

def remover_acentos_quebras_numeros(raw_corpus):
    raw_corpus = raw_corpus.str.replace('[\r\n]',' ')
    pt_stopwords = set([unidecode.unidecode(i) for i in nltk.corpus.stopwords.words('portuguese')])    
    def preprocess(x, *stopwords_list):
        x = unidecode.unidecode(x.lower())
        x = [w for w in x.split() if w not in set(pt_stopwords)]
        return ' '.join(x)
    
    clean_text = raw_corpus.apply(preprocess, args = pt_stopwords)
    clean_text = clean_text.str.replace('\d+', '_NUM_')
    return clean_text
```

```python
expr = re.compile('(\\n\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\n\\n\\n.*TRIBUNAL DE JUSTI.*V[iI][sS][tT][oO][sS]|\\n\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\t\\n\\n\\nTRIBUNAL DE JUSTIÇA DO ESTADO DE SÃO PAULO\\n\\n.*Juiz.* de Direito:.+?\\n)', re.S)
lista = df['Teor'].apply(lambda x: True if expr.search(x) else False)
df = df[lista]
df['Teor_preprocessado'] = remover_acentos_quebras_numeros(df['Teor'].apply(lambda x: re.sub(expr, '', x)))
df.head()
```

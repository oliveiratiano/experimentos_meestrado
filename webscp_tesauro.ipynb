```python
from bs4 import BeautifulSoup
import requests
import re
import sys
import time
import pandas as pd
from selenium import webdriver
from tqdm import tqdm_notebook
```

```python
driver = webdriver.Firefox(executable_path = 'D:\\programas\\geckodriver\\geckodriver.exe')
driver.get('http://www.stf.jus.br/portal/jurisprudencia/pesquisarVocabularioJuridico.asp')

alfabeto = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

lista_use = []
lista_up = []
lista_te = []
lista_tr = []
lista_tg = []
lista_cat = []
lista_nota = []
lista_termo = []


for letra in alfabeto:
    print('Raspando tela na letra '+letra)
    driver.execute_script('document.getElementById("passaletra").value = "'+letra.upper()+'"; document.getElementById("dropmsgform").submit();')
    time.sleep(10)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    elems = soup.find('tbody').find_all('tr')
    
    quant_termos_pagina = len(elems)
    i = 0
    
    while i < quant_termos_pagina:
        termo = elems[i].find('a').contents[0]
        lista_termo.append(termo)
        lista_use.append(" ")
        lista_up.append(" ")
        lista_te.append(" ")
        lista_tr.append(" ")
        lista_tg.append(" ")
        lista_cat.append(" ")
        lista_nota.append(" ")
        j = 1
        try:
            while(elems[i+j].find('div') == None):
                itens = elems[i+j].find_all('td')            
                if itens != None:
                    tipo = itens[1].find('strong').contents[0]
                    subitens = itens[2].find_all('a')
                    concats = '/'
                    for subitem in subitens:
                        concats += subitem.contents[0] + '/'                   

                    if(tipo == 'USE'):  
                        lista_use.pop()
                        lista_use.append(concats)
                    elif(tipo == 'UP'):
                        lista_up.pop()
                        lista_up.append(concats)
                    elif(tipo == 'TE'):
                        lista_te.pop()
                        lista_te.append(concats)
                    elif(tipo == 'TR'):
                        lista_tr.pop()
                        lista_tr.append(concats)
                    elif(tipo == 'TG'):
                        lista_tg.pop()
                        lista_tg.append(concats)
                    elif(tipo == 'CAT'):
                        lista_cat.pop()
                        lista_cat.append(concats)
                    elif(tipo == 'NOTA'):
                        lista_nota.pop()
                        lista_nota.append(concats)             
                j+=1
        except:
            print ("Erro:", sys.exc_info()[0])
        i+=j
```
